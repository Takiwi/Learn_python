# K - Nearest Neighbors (KNN)

```python
import numpy as np
import pandas as pd
```

## Bai tap

Bài toán (KNN, K = 3, khoảng cách Euclid)
- Sử dụng KNN với K = 3 để dự đoán kích thước áo cho 2 người sau. Hãy dùng khoảng cách Euclid (tổng giá trị tuyệt đối hiệu theo từng chiều):
    - Một người có chiều cao 161 cm và cân nặng 61 kg.
    - Một người có chiều cao 164 cm và cân nặng 59 kg.

### Data
Chú thích: Dữ liệu gồm các cặp Chiều cao (cm), Cân nặng (kg) và nhãn kích thước áo (M hoặc L).

```python
url = "./data/T_Shirt_size.csv"

data = pd.read_csv(url, encoding="utf-8")
data
```

## Câu 1

### 1. Tính khoảng cách

```python
target_height= 161
target_weight = 61

data["distance"] = np.sqrt(np.pow(data["Height (cm)"] - target_height, 2) + (np.pow(data["Weight (kg)"] - target_weight, 2)))
print(data)
```

### 2. Dự đoán (Với trọng số = 1)

```python
data_sorted = data.sort_values("distance")
print(data_sorted)
```

### K = 3 => Dự đoán là: 100 % là 1 người cao 161 cm và nặng 61 kg mặc size M 

## Câu 2

### 1. Tính khoảng cách

```python
target_height= 164
target_weight = 59

data["distance"] = np.sqrt(np.pow(data["Height (cm)"] - target_height, 2) + (np.pow(data["Weight (kg)"] - target_weight, 2)))
print(data)
```

### 2. Dự đoán (Với trọng số khác 1)

```python
data_sorted = data.sort_values("distance")
data_sorted["Distance-weight"] = 1 / data_sorted["distance"] 
print(data_sorted)
```

```python
print(data_sorted.head(3))
```

### K = 3 => Dự đoán là 1 người cao 164 cm, nặng 54 kg thì size là M vì Distance-weight lớn nhất

## Tính KNN bằng thư viện scikit-learn

```python
from sklearn.neighbors import KNeighborsClassifier

# Chuẩn bị dữ liệu
X = data[["Height (cm)", "Weight (kg)"]]
y = data["T shirt size"]

# Khởi tạo mô hình
knn = KNeighborsClassifier(n_neighbors=3)

# Huấn luyện mô hình
knn.fit(X, y)

# Dữ liệu cần dự đoán
new_data = pd.DataFrame([[161, 61]], columns=['Height (cm)', 'Weight (kg)'])
prediction = knn.predict(new_data)

print(f"Dự đoán size áo cho người cao 161cm, nặng 61kg là: {prediction[0]}")
```
