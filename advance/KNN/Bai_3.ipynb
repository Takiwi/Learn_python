### 2. Xây dựng hàm tính khoảng cách (Distance Function)

KNN dựa trên khoảng cách. Hãy viết một hàm bằng **NumPy** để tính khoảng cách Euclidean giữa hai điểm A và B.


### 3. Tìm các láng giềng gần nhất (Find Neighbors)

Viết một hàm nhận vào các tham số:

* `training_data`: Tập dữ liệu đã biết nhãn.
* `test_instance`: Một mẫu dữ liệu mới cần dự đoán.
* `k`: Số lượng láng giềng.

**Logic bên trong:**

1. Tính khoảng cách từ `test_instance` đến **tất cả** các điểm trong `training_data`.
2. Sắp xếp các điểm theo khoảng cách tăng dần.
3. Lấy ra `k` điểm có khoảng cách nhỏ nhất.

### 4. Bỏ phiếu bầu (Voting)

Từ danh sách `k` láng giềng đã tìm được, hãy đếm xem loại hoa nào xuất hiện nhiều nhất. Nhãn đó chính là dự đoán của mô hình.

* *Gợi ý:* Sử dụng `df['column'].value_counts()` hoặc `collections.Counter` trong Python.

### 5. Đánh giá mô hình

```python
import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo
```

## Chuẩn bị dữ liệu

```python
# fetch dataset
iris = fetch_ucirepo(id=53)

# Xáo trộn data
data = iris.data.original.sample(frac=1, random_state=42).reset_index(drop=True)

# Tính toán vị trí cắt 80/20
train_size = int(0.8 * len(data))

# Chia data
train_set = data.iloc[:train_size]
test_set = data.iloc[train_size:]

# Tách đặc trưng (X) và nhãn (y) 
X_train = train_set.drop('class', axis=1).values
y_train = train_set['class'].values

X_test = test_set.drop('class', axis=1).values
y_test = test_set['class'].values
```

## Tính khoảng cách

```python
# Số điểm
k = 3

predictions = []

# Duyệt qua từng bông hoa trong tập Test
for X_test_row in X_test:
    # Tính khoảng cách
    distance = np.sqrt(np.sum((X_train - X_test_row)**2, axis=1))
    
    # Tìm k láng giềng gần nhất
    nearest_ids = np.argsort(distance)[:k]
    nearest_labels = y_train[nearest_ids]

    # Lấy nhãn xuất hiện nhiều nhất
    pred = pd.Series(nearest_labels).mode()[0]
    predictions.append(pred)
```

## Kiểm tra độ chính xác

```python
# Cuối cùng, so sánh predictions với y_test
accuracy = np.mean(predictions == y_test)
print(f"Độ chính xác của mô hình: {accuracy * 100}%") 
```
