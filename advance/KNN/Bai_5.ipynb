## KNN imputer

Dựa trên file dataset **T_Shirt_size.csv"**, xóa vài ô và dùng thuật toán KNN imputer để điền lại và đồng thời đánh giá độ chính xác giá trị đoán với giá trị ban đầu. 

```python
import pandas as pd
import numpy as np
```

## Tạo dataset

```python
# Đọc dataset
T_Shirt_size = pd.read_csv("./data/T_Shirt_size.csv")
T_Shirt_size.head()

# Số lượng ô muốn xóa
n_missing = 5

# Chọn cột muốn xóa ô dữ liệu
cols_to_mask = ['Height (cm)', 'Weight (kg)']

for _ in range(n_missing):
    row_idx = np.random.randint(0, len(T_Shirt_size))
    col_name = np.random.choice(cols_to_mask)
    T_Shirt_size.loc[row_idx, col_name] = np.nan

# xử lý cột T shirt size bằng kỹ thuật One-hot Encoding
T_Shirt_size['is_M_size'] = np.where(T_Shirt_size['T shirt size'] == "M", 1, 0)
T_Shirt_size['is_L_size'] = np.where(T_Shirt_size['T shirt size'] == "L", 1, 0)
T_Shirt_size = T_Shirt_size.drop("T shirt size", axis=1)

data = T_Shirt_size
data
```

## Tính khoảng cách

```python
X = data[["Weight (kg)", "is_M_size", "is_L_size"]]
y = data["Height (cm)"]

def nan_euclidean_dist(a, b):
    n = len(a)

    # mask cho các cột chung
    mask = ~np.isnan(a) & ~np.isnan(b)
    n_common = np.sum(mask)

    # Nếu không có cột chung nào, trả về vô cực hoặc giá trị mặc định
    if n_common == 0:
        return np.inf
    
    # Tính tổng số bình phương khoảng cách trên các cột chung
    sum_sq = np.sum((a[mask] - b[mask])**2)

    # Áp dụng công thức phóng đại
    dist = np.sqrt((n / n_common) * sum_sq)
    return dist

# dist = nan_euclidean_dist(X, b)
```

```python
nan_value = X.iloc[1]
distance = []
for index in range(len(y)):
    if index == 1: continue
    distance.append(nan_euclidean_dist(nan_value, X.iloc[index]))

dis = np.array(distance)
indices = np.argsort(distance)
k_nearest_indices = indices[:3]
prediction = y[k_nearest_indices].sum() / 3
prediction
```
