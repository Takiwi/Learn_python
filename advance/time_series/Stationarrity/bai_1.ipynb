# Time series Stationarity 

## What is Stationarity?

A time series is stationary if it does not exhibit any long term trends or obvious seasonality.

It has:

- A constant variance through time
- A constant mean through time
- The statistical properties of the time series do not change

### import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# read in the data
url = '../data/daily-minimum-temperatures.csv'
df = pd.read_csv(url)
df.head()
```

### Preprocessing

```python
df['Date'] = pd.to_datetime(df['Date']) # Chuyen tu string sang datetime
df = df.sort_values('Date') # sap xep dataset theo thoi gian
df['Daily minimum temperatures'].unique()
```

cot <b>Daily minimum temperatures</b> co 1 so dong sai format co dau cham ? o truoc, nen can format lai

```python
def temperatures_format(text):
    if(text[0] == '?'): return text[1:]
    return text

df['Daily minimum temperatures'] = df['Daily minimum temperatures'].apply(temperatures_format).astype(float)    # Doi type tu string sang float
```

```python
df.info()
```

#### Hau nhu data tu real world deu la khong Stationarity vay nen ta can lam 1 vai buoc de dataset co tinh Stationarity

### Logarithm Transform
Logarithm Transform la ta lay $ ln(y_t) $ de nen cac du lieu ve, lam sao cho phuong sai cua dataset on dinh
$$ Y_t = log(X_t) $$

```python
def log_transform(value):
    if value > 0: return np.log(value) 
    return value

df['temperatures_log'] = df['Daily minimum temperatures'].apply(log_transform)
df.head()
```

### Differencing
Differencing la lay vi phan, phuong phap nay giup loai bo <b>Trend</b> bang cach lay gia tri hien tai tru cho gia tri truoc do
$$ Y_t = X_t - X_{t-1} $$
- $ Y_t $: Gia tri hien tai
- $ X_t $: Thoi gian hien tai
- $ X_{t-1} $: Thoi gian truoc do

luu y: Vi cong thuc la lay gia tri hien tai tru cho gia tri truoc do nen dong dau tien cua dataset se bi NaN vay nen ta se hy sinh dong nay bang cach xoa no di.

```python
df['temperatures_diff'] = df['Daily minimum temperatures'].diff()
df = df.dropna() # Xoa dong dau tien
df.head()
```

### Stationarity Test
Bay gio ta co the kiem tra tinh Stationarity cua dataset bang: <b>ADF test (Augmented Dickey–Fuller)</b>

ADF la kiem tra 1 thu goi la <b>Unit Root (Nghiem don vi)</b>, neu:
- Co Unit Root, no khong Stationarity
- Khong co Unit Root, no khong Stationarity
