### Bài 9 – Matrix Operations

Cho: ```A = np.array([[1, 2], [3, 4]])```
    ```B = np.array([[5, 6], [7, 8]])```
1. Nhân ma trận A @ B
2. Tính định thức của A
3. Tìm ma trận nghịch đảo của A

- Lưu ý: A * B và A @ B là khác nhau:
    - A * B là nhân từng phần tử
    - A @ B là nhân ma trận

````python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
A @ B
np.linalg.det(A)
np.linalg.inv(A)
````

### Bài 10 – Giải hệ phương trình

Giải hệ: 2x + y = 5, x + y = 3

Cách 1: Ta đưa về dạng chuẩn: A . X = B rồi áp công thức để tính X: X = A^-1 . B
- Trong đó:
    - A – coefficient matrix (ma trận hệ số)
    - X – variable vector (vector ẩn)
    - B – constant vector (vector hằng số)

Cách 2: Dùng ```np.linalg.solve(A, B)```

````python
A = np.array([[2, 1], [1, 1]])
B = np.array([[5], [3]])
X = np.linalg.inv(A) @ B
print("Cach 1:", X)

X = np.linalg.solve(A, B)
print("Cach 2:", X)
````

### Bài 11 – Dữ liệu điểm

Cho mảng điểm: ```scores = np.array([6.5, 8.0, 4.5, 9.0, 7.5, 3.0])```
1. Tính điểm trung bình
2. Đếm số sinh viên đậu (>= 5)
3. Chuẩn hóa điểm về khoảng [0, 1]

Công thức chuẩn hóa cho mọi trường hợp:


````python
scores = np.array([6.5, 8.0, 4.5, 9.0, 7.5, 3.0])
scores.mean()
scores[scores >= 5].size
scores = (scores - scores.min()) / (scores.max() - scores.min())
print(scores)
````

### Bài 12 – Mini Data Analysis

1. Tạo mảng (5, 3) random (0 → 100)
2. Tìm dòng có tổng lớn nhất
3. Chuẩn hóa mỗi cột (column-wise normalization)

````python
A = np.random.randint(0, 101, (5, 3))
max_sum = 0
row = -1
for x in range(A.shape[0]):
    if max_sum < A[x].sum(): 
        max_sum = A[x].sum()
        row = x
print(A)
print("Dong co tong lon nhat la:", row)
````
